```python
select review, min(price), avg(price), max(price)

from 

(select price,case when no_of_reviews = 0 then "new"
            when  no_of_reviews >= 1  and no_of_reviews <= 5 then "rising"
            when no_of_reviews >= 6  and no_of_reviews <= 15 then "trending up"
            when no_of_reviews >= 16  and no_of_reviews <= 40 then "popular"
            else "hot"
        end as review 
from reviews) t

group by review;
```
